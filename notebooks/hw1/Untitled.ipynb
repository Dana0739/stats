# Task 2

Cell coordinates are unoformely distributed. Petri dish is round.

```python
import random

def simulation(n):
    dots = {}
    i = 0
    while i < n:
        x1 = random.uniform(0, 1)
        y1 = random.uniform(0, 1)
        if ((0.5 - x1)**2 + (0.5 - y1)**2 <= 0.25):
            dots[x1] = y1
            i += 1
    return dots

def estimateCellCountsInRectangle(x1, y1, x2, y2, dots):
    res = {}
    for key in dots.keys():
        if key >= x1 and key <= x2:
            val = dots[key]
            if val >= y1 and val <= y2:
                res[key] = val
    return len(res)

dots = simulation(1000)
estimateCellCountsInRectangle(0.2, 0.2, 1, 1, dots)
```

```python
import matplotlib.pyplot as plt

xs = dots.keys()
ys = [dots[i] for i in xs]
plt.scatter(xs, ys)
```

The MLE of Î» in Poisson Distribution is the sample mean accoarding to the presentation

This takes us to generating a sample, so I am making this experiment several times.

```python
import numpy as np

n = 100
sample = []

for i in range(n):
    dots = simulation(1000)
    sample.append(estimateCellCountsInRectangle(0.2, 0.2, 1, 1, dots))
np.mean(sample)
```
